# Feature engineering 


On veut faire une régression linéaire pour prédire le prix d'un bien immobilier à partir de ses caractéristiques. On dispose d'un dataset contenant plusieurs caractéristiques : size (mètre carré), orientation, etc. 

Charger le dataframe et calculer le min et le max de la colonne size

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import tree 
from sklearn.tree import DecisionTreeClassifier
import graphviz
import seaborn as sns
sns.set_style("whitegrid")
```

```python
#df3=pd.read_csv("house_price_data_3.csv")
df4=pd.read_csv("house_price_data_4.csv")
#df4.describe()
df4
```

Afficher l'histogramme des valeurs

```python
sns.distplot(df4[["size"]],bins=30)
```

Normaliser la colonne size avec la méthode min max scaler afficher l'histogramme (ne pas écraser la colonne size pour le moment)

```python
x_min, x_max = df4["size"].min(),df4["size"].max()
#print(x_min, x_max)

x=df4[["size"]]
#x
```

```python
normalized = ((x-x_min)/(x_max-x_min))
normalized
```

```python
plt.hist(normalized["size"])
plt.show()
```

Faire de même avec la méthode standard scaler

```python
from sklearn.preprocessing import StandardScaler
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html
#des fois on peut faire le fit dans un dataset et transformer sur un autre dataset qui n'esst pas le même d'ou 
#d'ou l'importance de séparer le fit et le transform
scaler = StandardScaler()
print(scaler.fit_transform(x))
StandardScaler(copy=True, with_mean=True, with_std=True)
plt.hist(scaler.fit_transform(x))
plt.show()
```

Afficher la colonne "orientation". Quelles remarques avez-vous à faire ? 

```python
y=df4[["Orientation"]]
y
```

La colonne orientation est une colonne catégorielle. En faire le one-hot encoding avec la bonne méthode en pandas (chercher sur votre moteur de recherche favori) 

```python
z=pd.get_dummies(y)
z
#drop=first pour enlever la première colonne
```

```python
data=pd.concat([df4,z], axis=1)
#data
data.drop("Orientation", axis=1)
```

Faire une régression linéaire à l'aide de la variable size normalisée et l'orientation sous forme de one hot encoding

```python
sorted_df=data.sort_values(by=['size'])
sorted_df
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
#dans fit il y a x= variables explicatives et y=La variable à prédire
#
x=data[["size","Orientation_Est","Orientation_Nord","Orientation_Nord-Est","Orientation_Ouest","Orientation_Sud-Ouest"]]
y=data["price"]

model=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(x,y)
model.fit(x_train,y_train)

x_test_sorted = x_test.sort_values(by=['size']) #on l'ordonne par ordre croissant pour pouvoir faire la droite de 
#regression linéaire dans le schéma

model_predicted=model.predict(x_test_sorted)
#model_predicted

X=x.sort_values(by=["size"])#on l'ordonne par ordre croissant pour pouvoir faire la droite de 
#regression linéaire dans le schéma
#X
Y=model.predict(X)
#Y
```

```python
#yy=y.sort_values(by=["price"])
#yy
```

```python
plt.scatter(df4["size"][z["Orientation_Est"] == 1],df4["price"][z["Orientation_Est"] == 1],c="red", label="Orientation_Est")
plt.scatter(df4["size"][z["Orientation_Nord"] == 1],df4["price"][z["Orientation_Nord"] == 1],c="blue", label="Orientation_Nord")
plt.scatter(df4["size"][z["Orientation_Nord-Est"] == 1],df4["price"][z["Orientation_Nord-Est"] == 1],c="black", label="Orientation_Nord-Est")
plt.scatter(df4["size"][z["Orientation_Ouest"] == 1],df4["price"][z["Orientation_Ouest"] == 1],c="purple", label="Orientation_Ouest")
plt.scatter(df4["size"][z["Orientation_Sud"] == 1],df4["price"][z["Orientation_Sud"] == 1],c="green", label="Orientation_Sud")
plt.scatter(df4["size"][z["Orientation_Sud-Ouest"] == 1],df4["price"][z["Orientation_Sud-Ouest"] == 1],c="orange", label="Orientation_Sud-Ouest")
plt.legend()
plt.plot(X["size"],Y)
plt.show()
```

Afficher le coefficient de détermination

```python
#calcul du R²
R=r2_score(y,Y)
R
#Le tri par ordre croissant rendra la valeur positive
#https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html
```
